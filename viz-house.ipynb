# determine whether houses with a waterfront view or without a waterfront view have more price outliers .

```python
sns.set()
sns.boxplot(x="waterfront", y="price", data=df)
```

# Ploting price and square footage of house apart from basement

```python
sns.relplot(x="sqft_above", y="price", data=df)
```

# determine if the feature sqft_above is negatively or positively correlated with price.

```python
price_vs_sqft_above= df[['price','sqft_above']]
price_vs_sqft_above
```

```python
sns.lmplot(x='sqft_above', y = 'price', data = price_vs_sqft_above, line_kws={'color': 'red'})
```

# We can understand from the previous plot that the sqft_above have posative effect on the price and the relationship is direct proportion.
