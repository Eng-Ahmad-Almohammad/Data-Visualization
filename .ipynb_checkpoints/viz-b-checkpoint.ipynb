# Thea mean of charges for each age

```python
mean_for_charges = df.groupby('age',as_index=False).mean()
mean_for_charges
```

# Ploting the ages with the charges

```python
sns.relplot(x="age", y="charges", data=df)
```

```python
sns.relplot(x='age', y = 'charges', data = mean_for_charges)
```

# determine if the feature age is negatively or positively correlated with charge.

```python
sns.lmplot(x='age', y = 'charges', data = mean_for_charges, line_kws={'color': 'red'})
```
